So, with respect to these weights the most similar(closer) users to our test users are found.

```python
import matplotlib.pyplot as plt
import numpy as numpy
import pandas as pd
import sklearn as sk
import sklearn.metrics.pairwise
from numpy import linalg as LA 
import math
import random 
import matplotlib.pyplot as plt 

# create dataframes by reading the .csv files
ratings_df = pd.read_csv(ratings_path, delimiter= ",")
movies_df = pd.read_csv(movies_path, delimiter= ",", encoding="Latin1")

# create a dataframe for the whole ratings_train dataset
# rows are the user_ids and columns are the movie_ids
#total_movie_ids = numpy.array(total_movies_ratings.columns)
#total_movies_ratings = pd.pivot_table(ratings_df ,values='rating',index='userId',columns='movieId')
users_to_movies_df = pd.pivot_table(ratings_df ,values='rating',index='userId',columns='movieId')

print(ratings_df)

# numpy array representation of dataframes
ratings_numpy = ratings_df.to_numpy()
movies_numpy = movies_df.to_numpy()




```

```python
# defined Kfold function to implement k-fold for the future cross validation work
# takes the total data as a dataframe and also the fold number k
# devides the total_matrix into train_matrix and test_matrix
# returns the start end point indexes of the folds as an array


def Kfold(total_df, k):
      
    shape = numpy.array(users_to_movies_df.shape)
    num_of_ratings = shape[0]  
    # find the fold size
    fold_size = int(num_of_ratings / k )
    # remaining must be added to the folds if shuffled_df cannot be divided to k
    remaining = num_of_ratings % k
  
  # hold the start and end points of the folds in fold_indexes array
    fold_indexes = []
    for i in range(k):
        sub_index = [i*fold_size, ((i+1)*fold_size)-1]
        fold_indexes.append(sub_index)    
    # if there is a remaining, add it at the last fold
    if(remaining):
        sub_index[1] += remaining
    
    return fold_indexes


fold_indexes = Kfold(users_to_movies_df, 10)

print("fold_indexes: ", fold_indexes)
```

```python
  
# create a matrix and a df which rows are the users and columns are the movies
# values are the ratings for each movie given by each user in the df
# replace each NaN value with 0 or average

def create_movies_ratings(df_with_NaNs, mode):
    
    # replace NaN values with 0
    if(mode == 1):
        movies_ratings = df_with_NaNs.fillna(0)
    
    # replace NaN values with the user average (average along x-axis)
    if(mode == 2):
        movies_ratings = df_with_NaNs.fillna(df_with_NaNs.mean(axis=0))
    
    # replace NaN values with the movie average (average along y-axis)
    if(mode == 3):
        movies_ratings = df_with_NaNs.fillna(df_with_NaNs.mean(axis=1))
        # some movies are not voted so replace the remaining NaNs with 0
        movies_ratings = movies_ratings.fillna(0)
    
      
    movies_ratings_numpy = movies_ratings.to_numpy()
  
    #   #check whether there are any NaN values left
    #   print(movies_ratings.isna())
    #   print(movies_ratings.columns[movies_ratings.isnull().any()])
    
    return movies_ratings, movies_ratings_numpy


  
    
```

```python
# create train and test matrices where rows are user_ids 
# and columns are movie_ids and values are ratings
# with using the shuffled ratings_numpy dataset
# and also the fold start and end indeces


def create_train_test(df, ind_arr, mode):
  
    # these arrays hold the test and train dataframes in the index order
    # for example for the test[0] df corresponding train df is train[0]
    test = []
    train = []
    for ind in ind_arr:    
        start = ind[0]
        end = ind[1]    
        test_df = df.iloc[start:end+1]   
        movies_ratings_df, movies_ratings_numpy = create_movies_ratings(test_df, mode)
        test.append(movies_ratings_df)

        if ind == 0:
            rest_of_df = df[end+2:]
        else :
            #concat first part and second part of dfs
            temp1 = df[:start]
            temp2 = df[end+1:]
            rest_of_df = pd.concat([temp1, temp2])
        
        movies_ratings_df, movies_ratings_numpy = create_movies_ratings(rest_of_df, mode)
        train.append(movies_ratings_df)   
    
    return test, train
        
test, train = create_train_test(users_to_movies_df, fold_indexes, 1)
```

```python

# calculates cosine similarities for the given two dataframes
def cosine_similarity(df1, df2):
  
    # get a list of movie_ids 
    test_column_labels = list(df1.columns)
    # get a list of user_ids
    test_row_labels = list(df1.index.values)
    
    # get a list of movie_ids 
    train_column_labels = list(df2.columns)
    # get a list of user_ids
    train_row_labels = list(df2.index.values)  
    
    # compute similarity manually 
    df1_numpy = df1.to_numpy()
    df2_numpy = df2.to_numpy()
    dot = numpy.dot(df1_numpy, numpy.transpose(df2_numpy))
    norm1 = LA.norm(df1_numpy)
    norm2 = LA.norm(df2_numpy)
    
    sim_matrix =( dot / (norm1 * norm2)) 
    
    # create a df 
    # take the transpose of the sim_matrix
    # so the sim_df will be train_row_labels X test_row_labels
    sim_df = pd.DataFrame(sim_matrix, columns = train_row_labels, index= test_row_labels)
    
    return  sim_matrix, sim_df
  

################################################################################  

# holds the similarity matrices for each test and train dataframe combinations
# manuel similarity matrix array
sim_matrices = []
# manuel similarity dataframe
sim_df = []
for i in range(len(test)):
    man, df = cosine_similarity(test[i], train[i])
    sim_matrices.append(man)
    sim_df.append(df)
  

```

```python
# similarity matrix  holds dataframes 
# for test users_ids X train users_ids
# these dataframes are passed to knn() one by one
# finds the top k similar train users with test users
# returns a list of dictionary as dictionary_list ( it is actually a list, holds dictionaries)
# in each position in the dictionary_list, there is a dictionary for each test fold
# in each test fold, a dictionary will be added to this dictionary_list

dictionary_list= []

def calculate_weights(k,df, kernel):
  
    weights = []
    id_list = []  
    
    for i in range(k):
        if(kernel == "harmonic series"):
            weights.append(1/(i+1)) 
        
        if(kernel == "self-defined"):
            weights.append(math.pow(0.5,i))
        
        if(kernel == "uniform"):
            weights.append(1)
        
        
        if(kernel == "inverse-distance"):      
            df_numpy = df.to_numpy()
            # weights are 1 / distance in inverse-distance 
            dist = 1 - df_numpy[i]
            w = 1.0 / df_numpy[i]
            id_list.append(w)
            if( i == k-1):
                sum_of_weights = sum( id_list)
                new_id_list = [x / sum_of_weights for x in id_list]
                weights.extend(new_id_list)
      
    return weights
    
################################################################################    

def knn(df, k, kernel):
  
    # get a list of user_id column
    column_labels = list(df.columns)
    # get a list of user_id rows
    row_labels = list(df.index.values)
    
    temp_dict = {}
    
    for label in row_labels:
        temp_list = []
        # .loc returns pandas.Series 
        row_df = df.loc[label, :]
        # sort pandas.Series
        sorted_df = row_df.sort_values(0, ascending=False)
        # take the top-k highest values
        top_k = sorted_df[:k]
    
        weights = calculate_weights(k, top_k, kernel)
        for v in range(len(top_k.values)):
            print("weights for %d closest one: "%(v+1), weights[v])      
            temp_list.append([(weights[v] * top_k.values[v]), top_k.index[v]])    
            print("top-k values: ", top_k.values)
            print()
        temp_dict.update({label : temp_list})
    dictionary_list.append(temp_dict)  
    return

for i in range(len(sim_df)):
    knn(sim_df[i], 5, "harmonic series")
```

```python
# calculates the mean absolut error between each item in list1 and list2
def calculate_mae(list1, list2):
  
    n = len(list1)
    for i in range(n):
        err = (abs(list1[i] - list2[i])) / n
    return err


###############################################################################
# here for each test user, calculate the mae between corresponding train user

# ground-truth ratings (train)
gt_ratings = []
# predicted ratings (test)
pre_ratings = []

for d in dictionary_list:
    sub_gt = []
    sub_pre = []
    for k,v in d.items():
        subsub_gt = []
        subsub_pre = []  
        for item in v:
            movie_id = item[1]
            rating_predicted = item[0]
            subsub_pre.append(rating_predicted)
            sub_df = ratings_df.loc[(ratings_df.userId == int(k)) & (ratings_df.movieId ==int(movie_id) )]
            if sub_df.empty == False:
                rating_gt = sub_df.to_numpy().ravel()[2]
                subsub_gt.append(rating_gt)
            else:
                rating_gt = 0
                subsub_gt.append(rating_gt)
        sub_gt.append(subsub_gt)
        sub_pre.append(subsub_pre)    
    pre_ratings.append(sub_pre)
    gt_ratings.append(sub_gt)
  
    
# holds all the errors for all the folds
total_err_list =[]
for i in range(len(pre_ratings)):
    fold_err = []
    for k in range(len(pre_ratings[i])):
        err = calculate_mae(pre_ratings[i][k], gt_ratings[i][k])
        fold_err.append(err)
    total_err_list.append(fold_err)
 


 ##############################################################################
# here are the Mean Absolut Error Values
# holds the errors for every test fold
folds_err = []

for errList in total_err_list:
    err = sum(errList) / len(errList)
    folds_err.append(err)

# this is the total error for whole program test folds
total_err = sum(folds_err) / len(folds_err)


print("test fold's MAE: ", folds_err)
print("whole MAE: ", total_err)
```
